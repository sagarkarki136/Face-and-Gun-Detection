```
imgList = open('imgList.txt', 'r')
linesOfImg=imgList.readlines()
info='/content/WIDER_val/images/'
len(linesOfImg)
```

```
# os.remove('/content/imgSize.txt')

if not os.path.exists('imgSize.txt'):
	i=0
	while i < len(linesOfImg):
		f=open('imgSize.txt','a')
		img_name=str(linesOfImg[i].rstrip('\n'))
		img_path=info+img_name
		nImg=np.array(Image.open(img_path))
		
		f.write(str(nImg.shape[1])+' ')#width
		f.write(str(nImg.shape[0])+'\n')#height
		i = i+1
```

```
imgSize = open('imgSize.txt', 'r')
imgSize_lines=imgSize.readlines()
i=0
j=0
r=0
while i < numOfLines:
    num_face=lines[i+1]
    #print(num_face)
    img_name = lines[i].split('/')
  
    img_name = img_name[1][:-5]
    s_s=imgSize_lines[j].split()
    imgH=float(s_s[1])
    imgW=float(s_s[0])


    for numface in range(int(num_face)):
      f=open('/content/labels'+'/'+str(img_name)+'.txt','a')
      s=lines[i+2+numface].split()
      #print(s)
      #print(s[0])
      x=float(s[0])
      y=float(s[1])
      w=float(s[2])
      h=float(s[3])
      if x <= 0.0:
        x = 0.001
      if y <= 0.0:
        y = 0.001
      if w <= 0.0:
        w = 0.001
      if h <= 0.0:
        h = 0.001

      cx=x+0.5*w
      cy=y+0.5*h
      vocX=cx/imgW
      vocY=cy/imgH
      vocW=w/imgW
      vocH=h/imgH
      
      if vocX <= 0.0:
        print('x'+str(img_name))
      if vocY <= 0.0:
        print('y'+str(img_name))
      if vocW <= 0.0:
        print('w'+str(img_name))
      if vocH <= 0.0:
        print('h'+str(img_name))

    f.write('0'+' ')
    f.write(str(vocX)+' ')
    f.write(str(vocY)+' ')
    f.write(str(vocW)+' ')
    f.write(str(vocH)+' ')			
    f.write('\n')
  
    if int(num_face)==0:
      i=i+3
    else:
      i=i+int(num_face)+2

  
    print(j)
    j=j+1
    if j==3180:
      break

    
 
```
